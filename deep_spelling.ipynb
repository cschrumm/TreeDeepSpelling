```python
import random
import itertools
```

```python
# the alphabet.
alpha = ['a','b','c','d','e','f','g', \
        'h','i','j','k','l','m','n','o','p','q','r', \
        's','t','u','v','w','x','y','z']
# all combination of two
dbl = [x[0]+x[1] for x in itertools.product(alpha,alpha)]
# all combination of three
trpl = [x[0]+x[1]+x[2] for x in itertools.product(alpha,alpha,alpha)]
ln = len(alpha)
ky = {}
lln2 = len(dbl)
ky2 = {}
ll3 = len(trpl)
ky3 = {}

for i in alpha:
    ky[i] = 1.0
    
for i in dbl:
    ky2[i] = 1.0
    
for i in trpl:
    ky3[i] = 1.0
```

```python
brk = 0
# get the cound of the number of singles,double and triple character.
with open("words_alpha.txt") as f:
    for line in f:
        line = line.strip()
        
        for c in line:
            ky[c] = ky[c] + 1
            
        if len(line) > 2:
            for i in range(len(line)-1):
                pr = line[i:i+2]
                ky2[pr] = ky2[pr] + 1
        brk = brk + 1
        
        if len(line) > 3:
            for i in range(len(line)-2):
                pr = line[i:i+3]
                #print(pr)
                ky3[pr] = ky3[pr] + 1
                
# normalize the counts.
ln = sum([x[1] for x in ky.items()])
for k,i in ky.items():
    ky[k] = (i*1.0)/ ln
    
ln = sum([x[1] for x in ky2.items()])
    
for k,i in ky2.items():
    ky2[k] = (i*1.0)/ ln # len(ky2.items())

ln = sum([x[1] for x in ky3.items()])

for k,i in ky3.items():
    ky3[k] = (i*1.0)/ln
```
